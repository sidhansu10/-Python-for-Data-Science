# 1.Filter dataFrame based on a condition

```python
# find which rows have "Apple" in the "Company" column
df_laptops["Company"]=="Apple"
```

```python

# filter dataframe based on a condition
df_laptops[df_laptops['Company']=="Apple"]
```

```python
# verify which companies are in the "Company" column with value_counts()
df_laptops[df_laptops['Company']=="Apple"].value_counts('Company')
```

# 2. Create Column based on 1 conditions:
## np.where()

```python
import numpy as np
```

```python
# create an array based on price tiers
np.where(df_laptops['Price_euros'] > 2000 , 'Expensive' , 'Cheap')
```

```python
# set it to a new column
df_laptops['Price_tier'] = np.where(df_laptops['Price_euros'] > 2000 , 'Expensive' , 'Cheap')
```

```python
# show first five rows
df_laptops.head()
```

```python
# count values in price_tier column
df_laptops.value_counts('Price_tier')
```

# 3.Filter DataFrame based on multiple condition

```python
# find apple laptops that cost more than 2000 euros
(df_laptops['Company']=='Apple') & (df_laptops['Price_euros']>2000)
```

# 3.1 &(and)

```python
# filter dataframe based on multiple conditions
df_laptops[(df_laptops['Company']=='Apple') & (df_laptops['Price_euros']>2000)]
```

# 3.2 |(or)

```python
# find laptops Apple or Dell laptops
df_laptops['Company']=='Apple'
df_laptops['Company']=='Dell'

(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')
```

```python
# filter dataframe based on multiple conditions
df_laptops[(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')]
df_laptops[(df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')].value_counts('Company')
```

```python
# find laptops from Apple or Dell that cost over 2000 euros
((df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')) & (df_laptops['Price_euros']>2000)
```

```python
# filteer dataframe based on multiple conditions
df_laptops[((df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')) & (df_laptops['Price_euros']>2000)]
```
